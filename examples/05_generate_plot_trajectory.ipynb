```python
import matplotlib as plt
```

```python
first_out = '../build/output.json'
# first_out = '../build/output-1.json'
second_out = '../build/output-6.json'
# second_out = '../build/output-6.json'
```

```python
import json 

with open(first_out, "r") as robotPosFile: 
    data = json.load(robotPosFile)


len(data)
```

```python
x = [x['O_T_EE'][12] for x in data]
y = [y['O_T_EE'][13] for y in data]
z = [z['O_T_EE'][14] for z in data]

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
first_out
# Create a new figure
fig = plt.figure()

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Plot the points
ax.scatter(x, y, z, c='r', marker='o')

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```
